```python
import pandas as pd
import numpy as np
import pickle
```

```python
feats = [pd.read_pickle("raw_feats.pkl")]
labels = [pd.read_pickle("raw_labels.pkl")]
```

```python
df = pd.merge(feats[0], labels[0],  how='inner', left_on=['speaker_id','sentence_id', 'frame_number'], right_on = ['speaker_id','sentence_id', 'frame_number'])
del feats
del labels
```

```python
df = df.drop("speaker_id", 1)
df = df.drop("sentence_id", 1)
df = df.drop("frame_number", 1)
df = df.drop("use_x", 1)
df = df.drop("use_y", 1)
df = df.drop("label_x", 1)
df.columns = ["feature_vector", "label"]
```

```python
df = [df]
```
